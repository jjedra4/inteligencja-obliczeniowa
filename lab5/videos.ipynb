```python
import base64, zipfile, gymnasium as gym, torch
from pathlib import Path
from gymnasium.wrappers import RecordVideo
from IPython.display import HTML, display
from stable_baselines3 import SAC

def load_from_dir(model_dir, device):
    d = Path(model_dir)
    z = d.parent / (d.name + ".zip")
    z.unlink(missing_ok=True)
    with zipfile.ZipFile(z, "w", strict_timestamps=False) as zf:
        for f in d.iterdir():
            zf.write(f, f.name)
    return SAC.load(str(z), device=device)

_dev = "mps" if torch.backends.mps.is_available() else "cpu"
_model = load_from_dir("checkpoints/sac_lr_1em04/final", _dev)
_vdir = Path("videos/local/lr_1em04"); _vdir.mkdir(parents=True, exist_ok=True)

env = gym.make("LunarLanderContinuous-v3", render_mode="rgb_array")
env = RecordVideo(env, str(_vdir), episode_trigger=lambda _: True, disable_logger=True)
obs, _ = env.reset(seed=42); r_total = 0.0; done = False
while not done:
    a, _ = _model.predict(obs, deterministic=True)
    obs, r, term, trunc, _ = env.step(a); r_total += r; done = term or trunc
env.close()
print(f"SAC lr=1e-4  nagroda={r_total:.1f}")
mp4 = sorted(_vdir.glob("*.mp4"))[-1]
display(HTML(f'<video width="640" controls autoplay loop><source src="data:video/mp4;base64,{base64.b64encode(mp4.read_bytes()).decode()}" type="video/mp4"></video>'))
```
